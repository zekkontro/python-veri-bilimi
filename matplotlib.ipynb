```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10, 20, 30, 40, 50, 60, 70]
kiloListesi = [30, 40, 45, 62, 74, 81, 93]
```

```python
npYasListesi = np.array(yasListesi)
npKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(npYasListesi, npKiloListesi, "r")
plt.xlabel("Yaşlar")
plt.ylabel("Kilo Verileri")
plt.title("Yaşa göre kilo grafiği")
plt.show()
```

```python
numpyArray1 = np.linspace(0, 10, 20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1 ** 4
numpyArray3 = numpyArray1 ** 5
```

```python
plt.plot(numpyArray1, numpyArray2, "y*-")

plt.xlabel("Kuvveti 1 olan Sayılar")
plt.ylabel("Kuvveti 3 olan Sayılar")
plt.title("Kuvvetlere göre Sayılar")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(numpyArray1, numpyArray2, "r")
plt.subplot(1, 2, 2)
plt.plot(numpyArray2, numpyArray1, "b")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot([0, 2, 4, 6, 8, 10], [60, 52, 38, 87, 95, 77], "r")
figureAxes.set_xlabel("1. Sayı")
figureAxes.set_ylabel("2. Sayı")
figureAxes.set_title("Yeni Grafik")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1, 0.1, 0.7, 0.7])
eksen2 = figure2.add_axes([0.2, 0.4, 0.3, 0.3])

eksen1.plot(numpyArray1, numpyArray3, "g--")
eksen1.set_xlabel("X  Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Büyük Grafik")

eksen2.plot(numpyArray3, numpyArray2, "r--")
eksen2.set_xlabel("X  Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik")
```

```python
(myFigure, myExens) = plt.subplots(nrows=1, ncols=2)

for exen in myExens:
    exen.plot(numpyArray1, numpyArray3, "r")
    exen.set_xlabel("X Ekseni")
    exen.set_ylabel("Y Ekseni")
    
plt.tight_layout() ## Eksenler aarasına boşuk bırakır
```

```python
yeniFigure = plt.figure(dpi=100)
yeniEksen = yeniFigure.add_axes([0.3, 0.3, 0.8, 0.8])
yeniEksen.set_xlabel("Array 1")
yeniEksen.set_ylabel("Array 2")
yeniEksen.set_title("Karşılaştırma Grafiği")
yeniEksen.plot(numpyArray1, numpyArray2, "r--", label= "numpyArray ** 4")
yeniEksen.plot(numpyArray1, numpyArray3, "g--",label= "numpyArray ** 5")
yeniEksen.legend()
```

```python
yeniFigure.savefig("grafik1.png")
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyArray1, numpyArray3, color="#00FF00", alpha= 0.3, linewidth=5.0, linestyle="-.")
benimEksen.plot(numpyArray2, numpyArray3, linestyle=":", marker="o", markersize=8,markerfacecolor="red") ## marker veri noktalarını işaretler
```
